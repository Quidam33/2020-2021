```python
import numpy as np
from scipy. integrate import odeint
import matplotlib.pyplot as plt
import math

w2 = 6
tmax = 60
step = 0.05
y0 = [0.6, 1.6]

def W(y, t):
    y1, y2 = y
    return [y2, -w2*y1 ]

t = np.arange( 0, tmax, step)
w1 = odeint(W, y0, t)
y11 = w1[:,0]
y21 = w1[:,1]

fig = plt.figure(facecolor='white')
plt.plot(t, y11, linewidth=2)
plt.ylabel("x")
plt.xlabel("t")
plt.grid(True)
plt.show()
fig.savefig('01.png', dpi = 600)

fig2 = plt.figure(facecolor='white')
plt.plot(y11, y21, linewidth=2)
plt.ylabel("y")
plt.xlabel("x")
plt.grid(True)
plt.show()
fig2.savefig('02.png', dpi = 600)
```

```python
w2 = 12
g = 6

def f(t): 
    f = math.sin(6*t)
    return f

def W(y, t):
    y1, y2 = y
    return [y2, -w2*y1 - g*y2 + f(t) ]

t = np.arange( 0, tmax, step)
w1 = odeint(W, y0, t)
y11 = w1[:,0]
y21 = w1[:,1]

fig = plt.figure(facecolor='white')
plt.plot(t, y11, linewidth=2)
plt.ylabel("x")
plt.xlabel("t")
plt.grid(True)
plt.show()
fig.savefig('05.png', dpi = 600)

fig2 = plt.figure(facecolor='white')
plt.plot(y11, y21, linewidth=2)
plt.ylabel("y")
plt.xlabel("x")
plt.grid(True)
plt.show()
fig2.savefig('06.png', dpi = 600)
```
